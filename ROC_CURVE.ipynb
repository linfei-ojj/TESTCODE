```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
from sklearn.metrics import roc_curve, auc  
from sklearn import model_selection

fpr=[1,100/117,69/117,34/117,12/117,1/117,0]
tpr=[1,78/84,76/84,71/84,60/84,11/84,0]
fpr1=[1,1,115/117,44/117,10/117,0,0]
tpr1=[1,1,82/84,80/84,75/84,8/84,0]
fpr2=[1,63/117,53/117,36/117,12/117,2/117,0]
tpr2=[1,82/84,80/84,80/84,76/84,44/84,0]
fpr3=[1,99/117,74/117,31/117,7/117,0,0]
tpr3=[1,80/84,80/84,80/84,68/84,47/84,0]
fpr4=[1,1,108/117,31/117,10/117,0,0]
tpr4=[1,1,80/84,80/84,76/84,31/84,0]
roc_auc = auc(fpr,tpr)
roc_auc1 = auc(fpr1,tpr1)
roc_auc2 = auc(fpr2,tpr2)
roc_auc3 = auc(fpr3,tpr3)
roc_auc4 = auc(fpr4,tpr4)
```

```python
def plot_roc_curve(fpr, tpr):
    font1 = {'family' : 'Times New Roman',
    'weight' : 'normal',
    'size'   : 12,
    }
    font2 = {'family' : 'Times New Roman',
    'weight' : 'normal',
    'size'   : 9,
    }
    plt.subplots(figsize=(5,5))
    ax = plt.subplot()
    for label in (ax.get_xticklabels() + ax.get_yticklabels()):
        label.set_fontname('Times New Roman')
        label.set_fontsize(12)
    lw = 2
    roc_auc = auc(fpr,tpr)
    plt.plot(fpr, tpr, color='palegreen',marker='o',markersize = 6,lw=lw,linestyle='--',
             label='DOCTOR ROC curve (area = %0.3f)' % roc_auc)
    plt.plot(fpr1, tpr1, color='orange',marker='s',markersize = 6,lw=lw,linestyle='-.',
             label='SVM ROC curve (area = %0.3f)' % roc_auc1)
    plt.plot(fpr2, tpr2, color='hotpink',marker='^',markersize = 6,lw=lw,
             label='BAYES ROC curve (area = %0.3f)' % roc_auc2)
    plt.plot(fpr3, tpr3, color='red',marker='x',markersize = 6,lw=lw,
             label='FOREST ROC curve (area = %0.3f)' % roc_auc3)
    plt.plot(fpr4, tpr4, color='dodgerblue',marker='D',markersize = 6,lw=lw,linestyle=':',
             label='MEAN ROC curve (area = %0.3f)' % roc_auc4)

    plt.xlim([0, 1.0])
    plt.ylim([0, 1.05])
    plt.xlabel('False Positive Rate',font1)
    plt.ylabel('True Positive Rate',font1)
    plt.legend(loc="lower right",prop=font2)
    plt.savefig('XR_ROC_CURVE.png', dpi=1000, bbox_inches='tight')
    plt.show()  
    
plot_roc_curve(fpr, tpr) 
```
